```python
import numpy as np
import scipy.stats as sc

mu = 3.5
sigma = 0.5

# Lognormal Transformation
sigma = (np.log(1 + (sigma/mu)**2))**0.5
mu = np.log(mu) - 0.5 * sigma**2
print("mu", mu)
print("sigma", sigma)

# 5/95%
tn = 1.76
a = 0
n = 15
Xlower = np.exp(mu-tn*sigma*np.sqrt((1-a)+(1/n)))
Xupper = np.exp(mu+tn*sigma*np.sqrt((1-a)+(1/n)))
print("Lower", Xlower)
print("Upper", Xupper)
```

```python
import numpy as np
import scipy.stats as sc

berm_d = 3
berm_y = 18
berm_length = 0
dn = 4.5
ysat2 = (14.5 * 2.5 + 15 * 2 + berm_y * berm_d) / (dn + berm_d)
ysat = ysat2 - 1.645 * 0.05 * ysat2
print("y_sat", ysat)

k = 1.59e-3                    # Groter  ~ LN(7.52e-4, 3.76e-4)
k_h = 3.79e-7                  # Kleiner
D = 25.92                      # Groter  ~ LN(25, 0.5)
d = 3.66 + berm_d              # Kleiner ~ LN(4.5, 0.5), 3.66, SLOOT: 2.68
L_f = 16.60                    # Kleiner
B = 37.50                      # Kleiner
x_exit = B * 0.5 + berm_length # Kleiner SLOOT = 86.07
h = 7.65                       # Groter  ~ GIVEN
hp = 3.34                      # Kleiner ~ N(3.5, 0.1)
yw = 10                        # Groter
ys = 26.50                     # Kleiner ~ N(14.72, 0.736)
ich = 0.342                    # Kleiner
g = 9.81                       # DET
n = 0.25                       # DET 
phi = 37                       # DET
L = L_f + B + (x_exit - 0.5 * B)
v = 1.33e-6                    # DET
d70m = 2.08e-4                 # DET
d70 = 2.24e-4                  # Kleiner ~ LN(2.8e-4, 3.36e-5)
beq = 4.205940144

# UPLIFT
lambda_h = ((k * D * d) / k_h) ** 0.5
lambda_ = ((lambda_h) / (L_f + B + lambda_h)) * np.exp((0.5 * B - x_exit)/lambda_h)
phi_exit = hp + lambda_ * (h - hp)
dphi = phi_exit - hp
dphi_c = d * (ysat - yw) / yw
FoS = dphi_c / (1.2 * 1.2 * dphi)
bprov = np.absolute((np.log(FoS / 0.48) + 0.27 * beq) / 0.46)
Pfu = sc.norm.pdf(-bprov, loc=0, scale=1)

print("Uplift")
print("Phi_exit", phi_exit)
print("Lambda_h", lambda_h)
print("Resistance", dphi_c)
print("Load",dphi)
print("FoS", FoS)
print("Beta", bprov)
print("Pf", Pfu)



# HEAVE
i = lambda_ * (h - hp) / d
FoS = ich / (i * 1.2 * 1.2)
bprov = np.absolute((np.log(FoS / 0.37) + 0.3 * beq) / 0.48)
Pfh = sc.norm.pdf(-bprov, loc=0, scale=1)

print("\n\nHeave")
print("Resistance",ich)
print("Load",i)
print("FoS", round(FoS,3))
print("Beta", bprov)
print("Pf", Pfh)



# PIPING
F1 = n * ((ys / yw) - 1) * np.tan(np.radians(phi))
F2 = (d70m / (((v * k * L) / g) ** (1/3))) * ((d70 / d70m) ** 0.4)
F3 = 0.91 * (D / L) ** ((0.28 / ((D / L)**2.8 - 1)) + 0.04)
Hcp = F1 * F2 * F3 * L
FoS = (Hcp) / ((h - hp - 0.3*d) * (1.2 * 1.2))
bprov = np.absolute((np.log(FoS / 1.04) + 0.43 * beq) / 0.37)
Pfp = sc.norm.pdf(-bprov, loc=0, scale=1)

print("\n\nPiping")
print("Resistance", round(Hcp, 3))
print("Load", h - hp - 0.3*d)
print("FoS", FoS)
print("Beta", bprov)
print("Pf", Pfp)

print("\n\nUPPER Pf", np.min([Pfu, Pfh, Pfp]))
print("TOTAL FoS", (1.3e-5)/np.min([Pfu, Pfh, Pfp]), " > 1 (if good)")
```
